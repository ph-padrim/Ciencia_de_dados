```python
import pandas as pd

def solve_dataframe_q1_python(file_path):
    """
    Resolve a Questão 1 da seção DataFrame em Python, considerando a estrutura:
    primeira coluna 'Year', demais colunas nomes de países com PIB per capita.
    Lê o arquivo Excel, calcula a razão do PIB per capita de cada país
    em relação ao PIB per capita dos EUA e retorna um DataFrame com esses dados.

    """
    try:
        # Carregar a planilha
        df = pd.read_excel(file_path)

        # A primeira coluna é 'year', as demais são países
        # Definir 'year' como índice para facilitar o cálculo por ano
        df_indexed = df.set_index('year')

        # Verificar se 'United States' está entre as colunas de países
        if 'United States' not in df_indexed.columns:
            print("Erro: Coluna 'United States' não encontrada no arquivo Excel. Verifique o nome do país.")
            return None

        # Obter o PIB per capita dos EUA para todos os anos
        usa_gdp_per_capita = df_indexed['United States']

        # Calcular a razão do PIB per capita de cada país em relação ao dos EUA para cada ano
        # Excluir a coluna 'United States' do cálculo para evitar dividir por si mesmo
        countries_gdp = df_indexed.drop(columns=['United States'])

        # Usar broadcasting para dividir o PIB de cada país pelo PIB dos EUA para o mesmo ano
        ratio_df = countries_gdp.div(usa_gdp_per_capita, axis=0)

        # Adicionar a coluna 'United States' de volta ao DataFrame de razões com valor 1 (ou manter separada)
        # Para a questão, é mais útil ter a razão de todos os países, incluindo EUA (que seria 1)
        # Vamos recriar o DataFrame de razões incluindo todos os países e o ano
        full_ratio_df = df_indexed.div(usa_gdp_per_capita, axis=0)

        # Resetar o índice para 'Year' voltar a ser uma coluna
        full_ratio_df = full_ratio_df.reset_index()

        print("DataFrame com a razão do PIB per capita de cada país em relação aos EUA (por ano):")
        display(full_ratio_df)
        return full_ratio_df

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado.")
        return None
    except KeyError as e:
        print(f"Erro: Coluna não encontrada no arquivo Excel: {e}. Verifique os nomes das colunas e se 'Year' e 'United States' existem.")
        return None
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return None

if __name__ == "__main__":
    # Substitua pelo caminho correto do seu arquivo
    file_path = '/home/ph-padrim/Área de trabalho/Curso de Ciência dos Dados/ciencia-de-dados/Ciencia_de_dados/GDP_pc_Maddison_2020.xlsx'
    dataframe_result = solve_dataframe_q1_python(file_path)
    if dataframe_result is not None:
        # Você pode salvar o resultado em um CSV ou outro formato se desejar
        # dataframe_result.to_csv('gdp_ratio_vs_usa_python_adjusted.csv', index=False)
        pass
```

## Exercício 2

```python
def solve_dataframe_q2_python(file_path):
    """
    Resolve a Questão 2 da seção DataFrame em Python.
    Identifica países que, em pelo menos um ano, tiveram a razão do PIB per capita
    em relação aos EUA acima de 0.5.

    """
    try:
        # Carregar a planilha
        df = pd.read_excel(file_path)

        # A primeira coluna é 'Year', as demais são países
        df_indexed = df.set_index('year')

        if 'United States' not in df_indexed.columns:
            print("Erro: Coluna 'United States' não encontrada no arquivo Excel. Verifique o nome do país.")
            return []

        usa_gdp_per_capita = df_indexed['United States']

        # Calcular a razão do PIB per capita de cada país em relação ao dos EUA para cada ano
        # Excluir a coluna 'United States' do cálculo para evitar dividir por si mesmo
        countries_gdp = df_indexed.drop(columns=['United States'])

        # Usar broadcasting para dividir o PIB de cada país pelo PIB dos EUA para o mesmo ano
        ratio_df = countries_gdp.div(usa_gdp_per_capita, axis=0)

        # Encontrar países onde a razão é maior que 0.5 em pelo menos um ano
        # Usamos .any(axis=0) para verificar se há algum True em cada coluna (país)
        countries_above_threshold = ratio_df[ratio_df > 0.5].any(axis=0)

        # Filtrar apenas os países que atendem ao critério
        list_of_countries = countries_above_threshold[countries_above_threshold].index.tolist()

        print("Países com razão de PIB per capita em relação aos EUA acima de 0.5 em pelo menos um ano:")
        for country in list_of_countries:
            print(f"- {country}")
        
        return list_of_countries

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado.")
        return []
    except KeyError as e:
        print(f"Erro: Coluna não encontrada no arquivo Excel: {e}. Verifique os nomes das colunas e se 'Year' e 'United States' existem.")
        return []
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return []

if __name__ == "__main__":
    file_path = '/home/ph-padrim/Área de trabalho/Curso de Ciência dos Dados/ciencia-de-dados/Ciencia_de_dados/GDP_pc_Maddison_2020.xlsx'
    countries = solve_dataframe_q2_python(file_path)
    if countries:
        print(f"\nTotal de países encontrados: {len(countries)}")
```

## Exercício 3


```python
import pandas as pd

def solve_dataframe_q3_python(file_path):
    """
    Resolve a Questão 3 da seção DataFrame em Python.
    Calcula a razão média do PIB per capita para cada país, considerando apenas
    os anos em que a razão em relação aos EUA foi acima de 0.5.

    """
    try:
        # Carregar a planilha
        df = pd.read_excel(file_path)

        # A primeira coluna é 'Year', as demais são países
        df_indexed = df.set_index('year')

        if 'United States' not in df_indexed.columns:
            print("Erro: Coluna 'United States' não encontrada no arquivo Excel. Verifique o nome do país.")
            return None

        usa_gdp_per_capita = df_indexed['United States']

        # Transformar o DataFrame para o formato longo para facilitar o cálculo e filtragem
        # Primeiro, resetar o índice 'Year' para que ele se torne uma coluna novamente
        df_long = df_indexed.reset_index()
        
        # Usar melt para transformar as colunas de países em linhas
        df_long = df_long.melt(id_vars=['year'], var_name='Country', value_name='GDP_per_Capita')

        # Juntar com o PIB per capita dos EUA para cada ano
        usa_gdp_df = df_long[df_long['Country'] == 'United States'][['year', 'GDP_per_Capita']]
        usa_gdp_df = usa_gdp_df.rename(columns={'GDP_per_Capita': 'USA_GDP_per_Capita'})

        df_merged = pd.merge(df_long, usa_gdp_df, on='year', how='left')

        # Calcular a razão do PIB per capita de cada país em relação ao dos EUA
        df_merged['Ratio_vs_USA'] = df_merged['GDP_per_Capita'] / df_merged['USA_GDP_per_Capita']

        # Filtrar apenas os anos em que a razão foi acima de 0.5
        df_filtered = df_merged[df_merged['Ratio_vs_USA'] > 0.5]
        
        # Excluir 'United States' do cálculo da média, pois a razão dele consigo mesmo é sempre 1
        df_filtered = df_filtered[df_filtered['Country'] != 'United States']

        # Calcular a razão média para cada país
        average_ratio_per_country = df_filtered.groupby('Country')['Ratio_vs_USA'].mean().reset_index()
        average_ratio_per_country = average_ratio_per_country.rename(columns={'Ratio_vs_USA': 'Average_Ratio_above_0.5'})

        print("Razão média do PIB per capita em relação aos EUA (apenas anos com razão > 0.5):\n")
        print(average_ratio_per_country.head(10))
        return average_ratio_per_country

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado.")
        return None
    except KeyError as e:
        print(f"Erro: Coluna não encontrada no arquivo Excel: {e}. Verifique os nomes das colunas e se 'Year' e 'United States' existem.")
        return None
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return None

if __name__ == "__main__":
    file_path = '/home/ph-padrim/Área de trabalho/Curso de Ciência dos Dados/ciencia-de-dados/Ciencia_de_dados/GDP_pc_Maddison_2020.xlsx'
    result_df = solve_dataframe_q3_python(file_path)
    if result_df is not None:
        # Você pode salvar o resultado em um CSV ou outro formato se desejar
        # result_df.to_csv('gdp_average_ratio_above_0_5_python.csv', index=False)
        pass
```

```python
import pandas as pd

def solve_dataframe_q4_python(file_path):
    """
    Resolve a Questão 4 da seção DataFrame em Python.
    Gera o quantitativo de publicações por tipo de documento e ano de publicação,
    ordenado de forma decrescente pelo número de publicações, usando a planilha
    Scholar_Amostra.xlsx.

    """
    try:
        # Carregar a planilha
        df = pd.read_excel(file_path)

        # As colunas para tipo de documento e ano de publicação são 'PUB_TYPE' e 'PUB_YEAR'
        required_columns = ['PUB_TYPE', 'PUB_YEAR']
        if not all(col in df.columns for col in required_columns):
            print(f"Erro: Uma ou mais colunas necessárias ({required_columns}) não foram encontradas no arquivo Excel. Verifique os nomes das colunas.")
            print(f"Colunas disponíveis: {df.columns.tolist()}")
            return None

        # Gerar o quantitativo de publicações por tipo de documento e ano de publicação
        publication_counts = df.groupby(['PUB_TYPE', 'PUB_YEAR']).size().reset_index(name='Numero_de_Publicacoes')

        # Ordenar por número de publicações em ordem decrescente
        publication_counts_sorted = publication_counts.sort_values(by='Numero_de_Publicacoes', ascending=False)

        print("Quantitativo de publicações por tipo de documento e ano (ordenado decrescentemente):\n")
        print(publication_counts_sorted.head(20))
        return publication_counts_sorted

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado.")
        return None
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return None

if __name__ == "__main__":
    file_path = '/home/ph-padrim/Área de trabalho/Curso de Ciência dos Dados/ciencia-de-dados/Ciencia_de_dados/Scholar_Amostra.xlsx'
    result_df = solve_dataframe_q4_python(file_path)
    if result_df is not None:
        # Você pode salvar o resultado em um CSV ou outro formato se desejar
        # result_df.to_csv('publication_counts_python.csv', index=False)
        pass
```

```python
import pandas as pd

def solve_dataframe_q5_python(file_path):
    """
    Resolve a Questão 5 da seção DataFrame em Python.
    Calcula o número médio de citações das publicações por tipo de documento,
    usando a planilha Scholar_Amostra.xlsx.

    """
    try:
        # Carregar a planilha
        df = pd.read_excel(file_path)

        # As colunas para tipo de documento e número de citações são 'PUB_TYPE' e 'NUM_CITAT'
        required_columns = ['PUB_TYPE', 'NUM_CITATIONS']
        if not all(col in df.columns for col in required_columns):
            print(f"Erro: Uma ou mais colunas necessárias ({required_columns}) não foram encontradas no arquivo Excel. Verifique os nomes das colunas.")
            print(f"Colunas disponíveis: {df.columns.tolist()}")
            return None

        # Calcular o número médio de citações por tipo de documento
        average_citations = df.groupby('PUB_TYPE')['NUM_CITATIONS'].mean().reset_index()
        average_citations = average_citations.rename(columns={'NUM_CITATIONS': 'Media_de_Citacoes'})

        # Ordenar pelo número médio de citações (opcional, mas útil para visualização)
        average_citations_sorted = average_citations.sort_values(by='Media_de_Citacoes', ascending=False)

        print("Número médio de citações por tipo de documento:\n")
        print(average_citations_sorted)
        return average_citations_sorted

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado.")
        return None
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return None

if __name__ == "__main__":
    file_path = '/home/ph-padrim/Área de trabalho/Curso de Ciência dos Dados/ciencia-de-dados/Ciencia_de_dados/Scholar_Amostra.xlsx'
    result_df = solve_dataframe_q5_python(file_path)
    if result_df is not None:
        # Você pode salvar o resultado em um CSV ou outro formato se desejar
        # result_df.to_csv('average_citations_python.csv', index=False)
        pass
```

```python
import pandas as pd

def solve_dataframe_q6_python(file_path, population_column='VAR01'):
    """
    Resolve a Questão 6 da seção DataFrame em Python.
    Categoriza a população municipal em cinco categorias e conta o número de municípios
    em cada categoria para os municípios da região Sul do Brasil.
    
    """
    try:
        # Carregar a planilha
        df = pd.read_excel(file_path)

        # Verificar se as colunas necessárias existem
        required_columns = ['UF', population_column]
        if not all(col in df.columns for col in required_columns):
            print(f"Erro: Uma ou mais colunas necessárias ({required_columns}) não foram encontradas no arquivo Excel. Verifique os nomes das colunas.")
            print(f"Colunas disponíveis: {df.columns.tolist()}")
            return None

        # Filtrar municípios da região Sul
        south_region_states = ['PR', 'SC', 'RS']
        df_south = df[df['UF'].isin(south_region_states)].copy()

        if df_south.empty:
            print("Nenhum município encontrado para a Região Sul com os estados especificados.")
            return None

        # Remover valores nulos na coluna de população antes de categorizar
        df_south.dropna(subset=[population_column], inplace=True)

        if df_south.empty:
            print(f"Nenhum dado de população válido na coluna {population_column} para a Região Sul.")
            return None

        # Definir os intervalos de população fixos conforme solicitado
        bins = [0, 10000, 50000, 250000, 500000, df_south[population_column].max() + 1] # +1 para incluir o valor máximo
        labels = ["Muito Baixa", "Baixa", "Média", "Alta", "Muito Alta"]
        
        # Criar as categorias usando pd.cut com os limites fixos
        df_south['Categoria_Populacao'] = pd.cut(df_south[population_column],
                                                bins=bins,
                                                labels=labels,
                                                include_lowest=True,
                                                right=False) # right=False para intervalos [a, b)

        # Gerar a legenda com os intervalos de população reais
        legend_text = "\nLegenda das Categorias de População (Intervalos de População Real):\n"
        legend_text += f"- Muito Baixa: {bins[0]} - {bins[1]}\n"
        legend_text += f"- Baixa: {bins[1]+1} - {bins[2]}\n"
        legend_text += f"- Média: {bins[2]+1} - {bins[3]}\n"
        legend_text += f"- Alta: {bins[3]+1} - {bins[4]}\n"
        legend_text += f"- Muito Alta: {bins[4]+1} - {bins[5]-1} (a maior população)\n"
        print(legend_text)

        # Contar o número de municípios em cada categoria
        category_counts = df_south['Categoria_Populacao'].value_counts().reset_index()
        category_counts.columns = ['Categoria_Populacao', 'Numero_de_Municipios']

        # Ordenar as categorias
        category_counts['Categoria_Populacao'] = pd.Categorical(category_counts['Categoria_Populacao'],
                                                                 categories=labels,
                                                                 ordered=True)
        category_counts = category_counts.sort_values('Categoria_Populacao')

        print("Contagem de municípios por categoria de população na Região Sul:\n")
        print(category_counts)
        return category_counts

    except FileNotFoundError:
        print(f"Erro: O arquivo {file_path} não foi encontrado.")
        return None
    except KeyError as e:
        print(f"Erro: Coluna não encontrada no arquivo Excel: {e}. Verifique os nomes das colunas.")
        return None
    except Exception as e:
        print(f"Ocorreu um erro inesperado: {e}")
        return None

if __name__ == "__main__":
    file_path = '/home/ph-padrim/Área de trabalho/Curso de Ciência dos Dados/ciencia-de-dados/Ciencia_de_dados/REGIC2018_Cidades_v2.xlsx'
    population_column_name = 'VAR01'
    result_df = solve_dataframe_q6_python(file_path, population_column_name)
    if result_df is not None:
        # Você pode salvar o resultado em um CSV ou outro formato se desejar
        # result_df.to_csv(\'municipios_por_categoria_populacao_sul_python.csv\', index=False)
        pass
```
